```python
# Converting timestamp to separate columns of year, month, day, hour
# here 'x' is just a variable used for iterating in a for loop

df["hour"] = [x.hour for x in pd.DatetimeIndex(df.datetime)]
df["day"] = [x.dayofweek for x in pd.DatetimeIndex(df.datetime)]
df["month"] = [x.month for x in pd.DatetimeIndex(df.datetime)]
df["year"] = [x.year for x in pd.DatetimeIndex(df.datetime)]
```

```python
# Since we just have 2 years in dataset, we can have 0 for 2011 and 1 for 2012
df['year'] = df['year'].map({2011:0, 2012:1})
```

```python
df.head()
```

```python
# Now extract dependent and independent variables

X, Y = df.iloc[:, 1:], df['count']
```

```python
X.drop('count', axis = 1)
```

```python
# axis denotes 1 for column, and inplace denotes the changed to be done in the dataset too
X.drop('count', axis = 1, inplace = True)
```

```python
X
```

```python
plt.scatter(x = df['casual']+df['registered'], y= df['count'])
plt.show()
```

```python
X
```

```python
# delete casual and registered user column
X.drop(['casual', 'registered'], axis = 1, inplace = True)
```

```python
X.fillna(X.mean(), inplace = True)
Y.fillna(Y.mean(), inplace = True)

X
```

```python
# Applying machine learning model

# splitting the training dataset 70:30

from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
#x_test.fillna(x_test.mean(), inplace = True)
```

```python
# Features on larger scales can unduly influence the model. We want features on a similar scale.
# Scikit's preprocessing provides us with StandardScaler package to scale our data.


# we apply fit_transform on training dataset
# and just transform on test dataset, because fit, fits our dataset to the model

from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
x_train_scl = scl.fit_transform(x_train)
x_test_scl = scl.transform(x_test)
```

```python
#np.where(x_train.values >= np.finfo(np.float64).max)
```

```python
# Using Linear Regression Now

from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(x_train_scl, y_train)

# data is fitted to the model
# y_train_pred gets the prediction results on x_train
# y_test_pred gets the prediction results on x_test
y_train_pred = lreg.predict(x_train_scl)
y_test_pred = lreg.predict(x_test_scl)
```

```python
print(y_train.shape)
print(y_train_pred.shape)
```

```python
# Checking if y_test_pred contains any NaN

array_sum = np.sum(y_test_pred)
print(np.isnan(array_sum))
print(array_sum)


summ = np.sum(y_test)
print(np.isnan(summ))
print(summ)
```

```python
# Converting y_test_pred from float64 to int

#y_test_pred = y_test_pred.astype(np.int)

y_test_pred.dtype
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_squared_log_error

print("RMSE : train:", np.sqrt(mean_squared_error( y_train, y_train_pred)))
print("RMSE : test:", np.sqrt(mean_squared_error( y_test, y_test_pred)))

print("RMSLE: train: " , np.sqrt(mean_squared_log_error(abs(y_train), abs(y_train_pred))))
print("RMSLE: test: " , np.sqrt(mean_squared_log_error(abs(y_test), abs(y_test_pred))))



print('R2 train: %.3f' % r2_score(y_train, y_train_pred))
print('R2 test: %.3f' % r2_score(y_test, y_test_pred))
```

```python
# Trying Decision Tree Regressor

from sklearn.tree import DecisionTreeRegressor
dst = DecisionTreeRegressor()
dst.fit(x_train_scl, y_train)
y_train_pred2 = dst.predict(x_train_scl)
y_test_pred2 = dst.predict(x_test_scl)
```

```python
print("r2 score train", r2_score(y_train, y_train_pred2))
print("r2 score test", r2_score(y_test, y_test_pred2))

print("RMSE train :", np.sqrt(mean_squared_error((y_train), (y_train_pred2))))
print("RMSE test : ", np.sqrt(mean_squared_error((y_test), (y_test_pred2))))


print("RMSLE train :", np.sqrt(mean_squared_log_error(abs(y_train), abs(y_train_pred2))))
print("RMSLE test : ", np.sqrt(mean_squared_log_error(abs(y_test), abs(y_test_pred2))))

# Performance with Decision tree is better than Linear Regression
```
